```python
import pandas as pd
import numpy as np
```

```python
np.random.seed(42)
```

```python
num_records = 1000
```

```python
age = np.random.randint(18, 80, size=num_records)
income_level = np.random.normal(50000, 10000, size=num_records)
occupation_options = ['Engineer', 'Teacher', 'Doctor', 'Student', 'Unemployed', np.nan]
occupation = np.random.choice(occupation_options, size=num_records)
```

```python
mobile_app_usage = np.random.normal(0.6, 0.1, size=num_records)
fintech_disruption = np.random.normal(0, 0.1, size=num_records)
```

```python
compliance_costs = np.random.normal(500000, 100000, size=num_records)
regulatory_risks = np.random.normal(0, 100000, size=num_records)
regulatory_changes = np.random.choice([True, False, np.nan], size=num_records)
```

```python
nps = np.random.normal(7, 1, size=num_records)
service_quality = np.random.normal(0, 1, size=num_records)
```

```python
market_share = np.random.normal(0.2, 0.05, size=num_records)
brand_reputation = np.random.normal(0, 1, size=num_records)
```

```python
cost_income_ratio = np.random.normal(0.6, 0.1, size=num_records)
automation_initiatives = np.random.normal(0, 1, size=num_records)
```

```python
credit_risk = np.random.normal(0.5, 0.1, size=num_records)
capital_adequacy = np.random.normal(0.1, 0.05, size=num_records)
```

```python
strategic_alliances = np.random.normal(0, 1, size=num_records)
```

```python
data = pd.DataFrame({
    'Age': age,
    'Income Level': income_level,
    'Occupation': occupation,
    'Mobile App Usage': mobile_app_usage,
    'Fintech Disruption': fintech_disruption,
    'Compliance Costs': compliance_costs,
    'Regulatory Risks': regulatory_risks,
    'Regulatory Changes': regulatory_changes,
    'NPS': nps,
    'Service Quality': service_quality,
    'Market Share': market_share,
    'Brand Reputation': brand_reputation,
    'Cost-Income Ratio': cost_income_ratio,
    'Automation Initiatives': automation_initiatives,
    'Credit Risk': credit_risk,
    'Capital Adequacy': capital_adequacy,
    'Strategic Alliances': strategic_alliances
})
```

```python
data.to_csv('financial_data.csv', index=False)
```

```python
summary_stats = data.describe()
print(summary_stats)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(8, 6))
plt.hist(data['Age'], bins=20, color='skyblue', edgecolor='black', alpha=0.7)
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(data['Income Level'], data['Brand Reputation'], color='green', alpha=0.5)
plt.title('Income Level vs. Brand Reputation')
plt.xlabel('Income Level')
plt.ylabel('Brand Reputation')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
data.boxplot(column='Market Share', by='Occupation', figsize=(12, 8))
plt.title('Market Share by Occupation')
plt.xlabel('Occupation')
plt.ylabel('Market Share')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
num_missing_values = data.isnull().sum().sum()
num_unique_occupations = data['Occupation'].nunique()
print("\nNumber of missing values in the dataset:", num_missing_values)
print("Number of unique occupations:", num_unique_occupations)
```
